```python
x = np.zeros([len(df),10])
y = np.zeros([len(df),1])
```

```python
for i,row in df.iterrows():
#     if i <2000:
    for j,t in enumerate(row[36:46]):
#             print(type(x) , x)
        if type(t)== type("asdfa"):
#                 print(j)
            x[i][j] = 1
```

```python
x
```

```python
for i,row in df.iterrows():

    if type(row[7])== type("asdfa"):
        if row[7][0] == 'T':
            y[i] = 1
        if row[7][0] == 'A':
            y[i] = 2
        if row[7][0] == 'U':
            y[i] = 3
        if row[7][0] == 'L':
            y[i] = 4
```

```python
y
```

```python
s =0.7
split = int(s*len(df))

x_train = x[split:]
y_train = y[split:]

x_test = x[:split]
y_test = y[:split]
```

```python
model = np.linalg.inv(np.transpose(x_test).dot(x_test)).dot(np.transpose(x_test).dot(y_test))
```

```python
model.shape
```

```python
model
```

```python
pred=x_train.dot(model)
```

```python
pred = np.round(pred)
```

```python
np.std(pred-y_train)
```
